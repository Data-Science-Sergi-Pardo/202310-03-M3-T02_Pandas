## Exploració de les dades:

### Las variables que trobem al dataset són les següents:

- Year: 2008
- Month: 1-12
- DayofMonth: 1-31
- DayOfWeek: 1 (Dilluns) - 7 (Diumenge).
- DepTime: Hora de sortida real (local, hhmm).
- CRSDepTime: Hora de sortida programada (local, hhmm).
- ArrTime: Hora d'arribada real (local, hhmm).
- CRSArrTime: Hora d'arribada programada (local, hhmm).
- UniqueCarrier: Codi d'operador únic (Aerolínia). 
- FlightNum: Número de vol.
- TailNum: Número de cua de l'avió: és el registre de l'avió, un identificador únic de l'avió.
- ActualElapsedTime: Temps real de durada del vol en minuts. Resultat de sumar les variables: AirTime, TaxiIn i TaxiOut.
- CRSElapsedTime: Duració programada del vol. És el resultat de sumar ActualElapsedTime + ArrDelay (valors *negatius) + DepDelay.
- AirTime: Duració real del vol en minuts.
- ArrDelay: *Temps de retard a l’arribada, en minuts: En un vol es té en compte com "a temps" si el vol va operar en menys de 15 minuts després de l’hora programada que es mostra als Sistemes Computeritzats de Reserves (CRS) a les aerolínies. *Els valors negatius corresponen a una arribada abans de e temps.
- DepDelay: Retard a la sortida, en minuts.
- Origin: Origen IATA codi de l’aeroport.
- Dest: Desti IATA codi de l’aeroport.
- Distance: Distància en milles.
- Taxi-In: Taxi-In en minuts - És el temps que va des del moment de l’aterratge fins el moment del bloqueig.
- Taxi-Out: en minuts - És el temps que s’obté restant el temps des de que s’ extreuen les falques fins el temps d’enlairament real.
- Cancelled : Vol cancel·lat.
- CancellationCode: Motiu per la cancel·lació (A = operador, B = temps, C = NAS (Sistema Nacional de aviació, D = seguretat)
- Diverted: Desviat: 1 = yes, 0 = no
- CarrierDelay: en minuts. Retards associats a l’operador (aerolínies). Situacions que poden determinar un retard de l’operador són: Neteja de la aeronau, danys a l’aeronau, espera a l’arribada de passatgers o tripulació en connexió, equipatge, impacte amb aus, temps de càrrega, càtering, computadora, interrupció de l’equip de l’operador, legalitat de la tripulació (descans del pilot o assistent), danys per  mercaderies perilloses, inspecció d’enginyeria, aprovisionament de combustible, gestió de passatgers discapacitats, retard de la tripulació, servei de lavabo, manteniment, sobre vendes, servei d’aigua potable, trasllats o canvis en passatgers rebels, lentitud en l'embarcament o per seure, gestió de l’equipatge de mà, retards pel pes i l’equilibri.
- WeatherDelay: en minuts. Retards associats a les condicions climàtiques extremes o perilloses que es preveuen o es manifesten al moment de sortir, en ruta o al moment d’arribar.
- NASDelay: en minuts. El retard que està assignat al Sistema Nacional de l’Espai Aeri (NAS) pot incloure: condicions climàtiques no extremes, operacions aeroportuàries, volum de tràfic intens, control de tràfic aeri, etc.
- SecurityDelay: en minuts. El retards associats a la seguretat. Són causats per: l’evacuació d’una terminal o vestíbul, el reembarcament d’una aeronau degut a una violació de la  seguretat, equip de control inoperatiu i/o llargues cues a les àrees de control (més de 29 minuts).
- LateAircraftDelay: en minuts. Retard en l’arribada a l’aeroport com a conseqüència d’un retard en l’anterior aeroport. L’efecte dominó d’un retard anterior en els posteriors aeroports es coneix com a propagació del retard.

- Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...).

```python
# Als EUA la velocitat s'expressa en milles per hora mph i com ens demana la velocitat mitjana tindrem en compte tot el temps transcorregut i no només el temps de vol.

flights['AvgSpeed'] = flights['Distance'] / (flights['ActualElapsedTime']/60)

# Per crear la variable booleana si ha arribat tard o no tindrem en compte el criteri utilitzat per Kaggle, només es considera retard aquells vols que arrDelay > 15 min.

flights['Delay'] = 'No'
flights.loc[flights['ArrDelay'] > 15, 'Delay'] = 'Yes'

flights.head(5)
```

- Fes una taula de les aerolínies amb més endarreriments acumulats.

```python
print('Les aerolínies amb més endarreriment acumulat són:\n')

carrier_delay_total = flights.groupby('UniqueCarrier')['ArrDelay'].sum()
carrier_delay_total_sorted = carrier_delay_total.sort_values(ascending=False).reset_index()

carrier_delay_total_sorted
```

- Quins són els vols més llargs? I els més endarrerits? Busca les rutes més llargues i les que acumulen més retards.

```python
# Els vols més llargs, entenent per temps 

longest_flights = flights.sort_values(by='AirTime', ascending=False)

top_10_longest_flights = longest_flights[['UniqueCarrier', 'TailNum', 'AirTime', 'Distance', 'Origin', 'Dest']].head(10)

top_10_longest_flights
```

```python
# Els més endarrerits

most_delayed_flights = flights.sort_values(by='ArrDelay', ascending=False)

top_10_most_delayed_flights = most_delayed_flights[['UniqueCarrier', 'TailNum', 'ArrDelay', 'Distance', 'Origin', 'Dest']].head(10)

top_10_most_delayed_flights
```

```python
# Les rutes més llargues

flights['Route'] = flights[['Origin', 'Dest']].min(axis=1) + '-' + flights[['Origin', 'Dest']].max(axis=1)

unique_routes = flights[['Route', 'Distance']].drop_duplicates()

top_longest_routes = unique_routes.sort_values(by='Distance', ascending=False)

top_longest_routes.head(5)
```

- Els vols més llargs tant d'anada com de tornada són entre:
    - Honolulu i Newark (New Jersey)
    - Atlanta i Honolulu
    - Honolulu i O'Hare (Chicago)
    - Kona (Hawái) i O'Hare (Chicago)
    - Kahului (Hawái) i O'Hare (Chicago)

```python
# Les rutes que acumulen més retards.

unique_routes = flights[['Route', 'ArrDelay']].drop_duplicates()

top_longest_routes = unique_routes.sort_values(by='ArrDelay', ascending=False)

top_longest_routes.head(5)
```

- Els vols que acumulen més retrads tant d'anada com de tornada són entre:
    - Honolulu i Minneapolis
    - Charlotte (Carolina del Norte) i Minneapolis
    - Detroit i Florida
    - Dallas-Fort Worth i Arkansas
    - Bostón i Minneapolis

- Aporta allò que consideris rellevant.

```python
print('Retards acumulats segons el dia de la setmana:\n')
      
delay_per_day = flights.groupby('DayOfWeek')['ArrDelay'].sum()

delay_per_weekday = delay_per_day.sort_values(ascending=False).reset_index()

delay_per_weekday
```
